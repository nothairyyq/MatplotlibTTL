```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

## First Example
- Matplotlib 最基本的图像呈现在窗体中(figure)
- 每一个figure包含一个或多个axes
- `pyplot.subplots`是最基本创建figure和axes的方式
- `Axes.plot` 可以绘制折线图

```python
y=np.random.normal(2,1,10)
x=np.arange(10)
#创建一个包含一个axes的图像
fig,ax=plt.subplots()
p=ax.plot(x,y)
```

上述block最后显示出`[<matplotlib.lines.Line2D at 0x1f0b567dec8>]`
有三种method消除这段话
1. 在block后加;
2. 在block后加plt.show()
3. 绘图时将绘图赋值给一个变量

## Figure Concepts
- Figure: Top level(container),用来容纳所有绘图元素
- Axes: matplotlib's core, 一个figure可以包括一个或多个axes
- Axis: axes的下属层级，用于处理所有的坐标轴，网格有关的元素
- Tick: axis的下属层级，用于处理所有和刻度有关的元素

## Two Plot API
#### 1. Object-Oriented Style
显示创建figure和axes，在上面调用绘图方法

```python
x=np.linspace(0,2,100)
fig,ax=plt.subplots()
ax.plot(x,x,label='Linear')
ax.plot(x,x**2,label='Quadratic')
ax.plot(x,x**3,label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Plot example')
ax.legend();
```

#### 2. other style
依赖`pyplot`自动创建figure和axes

```python
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Plot example')
plt.legend();
```

### Template

```python
# Step1: 准备数据
x=np.linspace(0,2,100)
y=x**2

# Step2: 设置绘图样式
mpl.rc('lines',linewidth=4,linestyle='-.')

# Step3: 定义布局
fig,ax=plt.subplots()

# Step4: 绘制图像
ax.plot(x,y,label='Linear')

# Step5: 添加label, title, legend
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Plot example')
ax.legend();
```
