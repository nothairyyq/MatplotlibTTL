```python
import numpy as np
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

### 1. matplotlib 的 三层API
matplotlib的fundamental principal:
    用artist对象在canvas上render图形
1. 准备一块canvas
2. 准备好颜料，画笔等render
3. 作画

- `matplotlib.backend_bases.FigureCanvas`；
    绘图区，所有的图像都是在绘图区完成
- `matplotlib.backend_bases.Renderer`；
    渲染器，即画笔，控制如何在canvas上画图
- `matplotlib.artist.Artist`；
    具体的画图组件，调用renderer的接口在canvas上作图

### 2. Artist的分类
##### primitive
    primitive是基本要素
    包含需要在绘图区用到的标准图形对象。
    eg. Line2D(曲线)，text，rectangle,image ...

##### container
    container用来装基本要素
    包括图形figure，坐标系axes和坐标轴axis

### 2D Lines
曲线的绘制: `matplotlib.lines.Line2D`

#### Line Definition:
连接所有顶点的实线样式，也可以是每个顶点的标记。

#### 常用Parameters
- **xdata**:
    将要绘制line中点在x轴上的取值
    默认值：`range(1,len(ydata)+1)`
- **ydata**
    line中点在y轴上的取值
- **linewidth**
    线条宽度
- **linestyle**
    line pattern
- **color**
- **marker**
    点标记
- **markersize**
    标记的size

[Official guide](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html)

## 绘制lines
- plot方法绘制
- Line2D对象绘制

```python
x=range(0,100)
y1=np.random.normal(0,1,100)
y2=np.random.normal(1,2,100)
```

```python
# plot方法直接绘制
fig1,ax1=plt.subplots()
ax1.plot(x,y2)
ax1.plot(x,y1)
print(ax1.lines)
plt.show()
# matplotlib在底层创建了两个line2D对象

fig1,ax1=plt.subplots()
##显示创建Line2D对象
lines = [Line2D(x,y1),Line2D(x,y2,color='green')]
for line in lines:
    ax1.add_line(line) #将创建的line2D两个对象加入到子图中
ax1.set_xlim(0,100)
ax1.set_ylim(-5,10)
```

### errorbar绘制误差折线图
#### parameters
- x：需要绘制的line中点的在x轴上的取值

- y：需要绘制的line中点的在y轴上的取值

- yerr：指定y轴水平的误差

- xerr：指定x轴水平的误差

- fmt：指定折线图中某个点的颜色，形状，线条风格，例如‘co--’

- ecolor：指定error bar的颜色

- elinewidth：指定error bar的线条宽度

```python
fig=plt.figure()
x=np.arange(50)
y=2.5*np.sin(x/20*np.pi)
yerr=np.linspace(0.5,10,50)
plt.errorbar(x,y+2,yerr=yerr)
```
